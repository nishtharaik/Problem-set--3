Question 1: 
Introduction: Special thanks to: https://github.com/justmarkham for sharing the dataset and materials. Occupations Step 1. Import the necessary libraries Step 2. Import the dataset from this address. Step 3. Assign it to a variable called users Step 4. Discover what is the mean age per occupation Step 5. Discover the Male ratio per occupation and sort it from the most to the least Step 6. For each occupation, calculate the minimum and maximum ages Step 7. For each combination of occupation and sex, calculate the mean age Step 8. For each occupation present the percentage of women and men

```python
#Step 1: Importing the necessary files

import pandas as pd
```

```python

#Step 2 and 3: importing the dataset and assigning it to a variable called 'users'

users = pd.read_csv(r"https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user",sep="|")
users.head()
```

```python
#Step 4 : discovering the meand age per occupation
meanAgePerOccupation = users.groupby("occupation").age.mean()
meanAgePerOccupation 
```

```python
#Step 5 : discovering the male ratio per occupation and sorting it from the most to the least

males = users.where(users.gender=="M").groupby(["occupation","gender"]).gender.count()
total = users.groupby("occupation").gender.count()
maleRationPerOccupation = (males / total)*100
maleRationPerOccupation.sort_values(ascending=False)
```

```python
#Step 6 : calculating the minimum and maximum ages for each occupation
minAndMax = users.groupby(["occupation"]).age.agg(["min","max"])
minAndMax
```

```python
#Step 7 : calculating mean age for each combination of occupation and sex

meanAge = users.groupby(["occupation","gender"]).age.mean()
meanAge
```

```python
#Step 8 : presenting the percentage of women and men for each occupation

genderPercentage = users.groupby('occupation')['gender'].value_counts(normalize=True)*100
genderPercentage
```

Question 8:

Create a horizontal bar chart that compares the deadliest actors in Hollywood. Sort
the actors by their kill count and label each bar with the corresponding actor's name.
Don't forget to label your axes!

```python

import matplotlib.pyplot as plt
actors = pd.read_csv('actor_kill_counts.csv')
sortedData=actors.sort_values(by='Count',ascending=True)
sortedData.plot.barh(x='Actor', y='Count',color="maroon")
plt.xlabel("Number of Kills", fontsize=14)
plt.ylabel("Name of the Actors", fontsize=14)
plt.title("The Deadliest Actors in Hollywood", fontsize=21)
plt.show()
```

Question 9 :
Create a pie chart showing the fraction of all Roman Emperors that were
assassinated.
Make sure that the pie chart is an even circle, labels the categories, and shows the
percentage breakdown of the categories.

```python
romanData = pd.read_csv('roman-emperor-reigns.csv')
causeOfDeath = romanData.groupby("Cause_of_Death")['Emperor'].count()

causeOfDeath.plot.pie(autopct="%.1f%%", fontsize = 11, ylabel = '')
plt.title("Pie Chart Showing The Cause of Death of Roman Emperors", color = 'Orange', fontsize = 21)
```

Question 10
Create a scatter plot showing the relationship between the total revenue earned by
arcades and the number of Computer Science PhDs awarded in the U.S. between
2000 and 2009.
Don't forget to label your axes!
Color each dot according to its year.

```python
USData = pd.read_csv('arcade-revenue-vs-cs-doctorates.csv')
colors = ["Violet","Indigo","Blue","Green","Yellow","Orange","Red","Black","Gray","Pink"]

USData.plot.scatter(x='Total Arcade Revenue (billions)', y='Computer Science Doctorates Awarded (US)', color=colors, s=100)
plt.title('Relationship Between Revenue and Computer Science PHD Award', color = 'Orange', fontsize = 10)
plt.xlabel('Total Revenue of Arcades (billions)' , fontsize = 14)
plt.ylabel('No. Of Computer Science PHD Awarded in the U.S.', fontsize = 10)
```
